```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from utils import extract_frame
```

```python
frame = extract_frame('./data/problem_1.mp4', 350)
invGamma = 1.0 / 2.9
beta = 0.8
table = np.array([beta*((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype(np.uint8)
frame2 = cv2.LUT(frame, table)
frame2 = cv2.GaussianBlur(frame2, (11,11),2)
fig, ax = plt.subplots(2,2,figsize=(15,15))
plt.subplot(121)
plt.imshow(frame2)
plt.title("Gamma Correction")
plt.subplot(122)
plt.imshow(frame)
plt.title("Original")
plt.show()
```

```python
filepath = './data/problem_1.mp4'
invGamma = 1.0 / 2.2
beta = 0.7
table = np.array([beta*((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype(np.uint8)
cap = cv2.VideoCapture(filepath)
if (cap.isOpened()== False): 
        print("Error opening video stream or file")    
while(cap.isOpened()):
    try:
        ret, frame = cap.read()
        frame = cv2.LUT(frame,table)#np.array(2*frame**(1/1.2)).astype(np.uint8)
        cv2.imshow('frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    except:
        break
cv2.destroyAllWindows()
```
